# Practical: Forest Fire Data Analysis using PyHadoop (MapReduce)

##  Aim  
To integrate Python with Hadoop using MapReduce (mrjob) and perform data analysis on the Forest Fire dataset to calculate the **average burned area per month**.

---

##  Objectives  
- To understand Big Data processing using Hadoop  
- To learn Python integration with Hadoop using mrjob  
- To implement MapReduce programming model  
- To analyze Forest Fire dataset  
- To calculate average burned area month-wise  

---

##  Theory  

###  Hadoop  
Hadoop is an open-source framework used to store and process large datasets in a distributed environment using multiple systems.

###  MapReduce  
MapReduce is a programming model used for processing large data in two phases:

**Mapper:**  
Processes input and converts it into key-value pairs  
Example:  
(month, (area, 1))  

**Reducer:**  
Aggregates values and computes final output  
Example:  
Average = Total Area / Count  

---

###  PyHadoop (mrjob)  
- mrjob is a Python library used to implement MapReduce  
- It allows running Hadoop jobs using Python  
- Can run locally or on Hadoop cluster  

---

### Dataset Description  
Forest Fire dataset includes:  
- X, Y → Coordinates  
- Month, Day → Time of fire  
- Temperature, Humidity, Wind → Weather conditions  
- Rain → Rainfall  
- Area → Burned area (hectares)  

---

###  Working  

**Mapper Phase:**  
- Reads each record  
- Extracts month and burned area  
- Outputs: (month, (area, 1))  

**Reducer Phase:**  
- Calculates total area and count  
- Computes average burned area  

---

###  Output  
Average burned area for each month:  

jan → 0.1234  
feb → 0.5678  
...
